genererar två QuickStatements-filer
* Commons Structured Data → P4819 (SPA-ID på filens M-entity)
* Wikidata → P2488 (SPA-ID på personen)

```python
from datetime import datetime
start_time  = datetime.now()
print("Last run: ", start_time)
```

```python
import requests
import re
import time
from tqdm import tqdm

version = "v2"
API = "https://commons.wikimedia.org/w/api.php"
WD_API = "https://www.wikidata.org/w/api.php"

CATEGORY = "Category:Swedish_Portrait_Archive"

HEADERS = {
    "User-Agent": "SPA-sync-bot/1.0 (salgo60@msn.com)"
}

session = requests.Session()
session.headers.update(HEADERS)

SPA_REGEX = re.compile(r"portrattarkiv\.se/details/([A-Za-z0-9]+)", re.I)


def api_get(url, params):
    r = session.get(url, params=params, timeout=30)
    r.raise_for_status()
    return r.json()


def get_category_files():

    files = []
    cont = None

    while True:

        params = {
            "action": "query",
            "format": "json",
            "list": "categorymembers",
            "cmtitle": CATEGORY,
            "cmtype": "file",
            "cmlimit": 500
        }

        if cont:
            params["cmcontinue"] = cont

        data = api_get(API, params)

        files.extend(data["query"]["categorymembers"])

        if "continue" in data:
            cont = data["continue"]["cmcontinue"]
        else:
            break

    return files


def get_pages(titles):

    params = {
        "action": "query",
        "format": "json",
        "prop": "revisions",
        "rvprop": "content",
        "titles": "|".join(titles)
    }

    data = api_get(API, params)

    return data["query"]["pages"]


def get_commons_entities(mids):

    params = {
        "action": "wbgetentities",
        "format": "json",
        "ids": "|".join(mids)
    }

    data = api_get(API, params)

    return data["entities"]


def get_wd_entities(qids):

    params = {
        "action": "wbgetentities",
        "format": "json",
        "ids": "|".join(qids)
    }

    data = api_get(WD_API, params)

    return data["entities"]


def extract_spa(text):

    m = SPA_REGEX.search(text)

    if m:
        return m.group(1)

    return None


print("Loading category members…")

files = get_category_files()

print("files:", len(files))

qs_commons = []
qs_wd = []

BATCH = 50

stats = {
    "spa_found": 0,
    "commons_entity": 0,
    "depicts": 0,
    "wd_missing_spa": 0
}

for i in tqdm(range(0, len(files), BATCH), desc="Processing batches"):

    batch = files[i:i+BATCH]

    titles = [f["title"] for f in batch]

    pages = get_pages(titles)

    mids = []
    spa_map = {}

    for p in pages.values():

        text = p["revisions"][0]["*"]

        spa = extract_spa(text)

        if not spa:
            continue

        stats["spa_found"] += 1

        pageid = p["pageid"]
        mid = "M" + str(pageid)

        stats["commons_entity"] += 1

        mids.append(mid)
        spa_map[mid] = spa

    if not mids:
        continue

    commons_entities = get_commons_entities(mids)

    qids = []
    file_person_map = {}

    for mid, entity in commons_entities.items():

        claims = entity.get("claims", {})

        # kontrollera om SPA redan finns
        if "P4819" not in claims:
            qs_commons.append(f"{mid}|P4819|\"{spa_map[mid]}\"")

        if "P180" not in claims:
            continue

        for dep in claims["P180"]:

            wd = dep["mainsnak"]["datavalue"]["value"]["id"]

            stats["depicts"] += 1

            spa = spa_map[mid]

            file_person_map[wd] = spa

            qids.append(wd)

    if not qids:
        continue

    wd_entities = get_wd_entities(qids)

    for qid, entity in wd_entities.items():

        claims = entity.get("claims", {})

        if "P2488" in claims:
            continue

        spa = file_person_map[qid]

        stats["wd_missing_spa"] += 1

        qs_wd.append(f"{qid}|P2488|\"{spa}\"")

    time.sleep(0.2)


qs_common_spa_file = "quickstatement_commons_spa_" + version + ".txt"
with open(qs_common_spa_file, "w") as f:
    f.write("\n".join(qs_commons))

qs_wd_spa_file = "quickstatement_wikidata_spa_" + version + ".txt"
with open(qs_wd_spa_file, "w") as f:
    f.write("\n".join(qs_wd))

clean = []

print("\nDone")

print("Commons statements:", len(qs_commons))
print("Wikidata statements:", len(qs_wd))
print("Stats:", stats)
```

```python
clean = [] 

for line in qs_commons:
    spa = line.split('"')[1]
    if len(spa) >= 18:
        clean.append(line)

qs_wd_spa_file_clean = "quickstatement_wikidata_spa_clean_" + version + ".txt"

with open(qs_wd_spa_file_clean,"w") as f:
    f.write("\n".join(clean))
```
